```python
import folium
import pandas as pd

df = pd.read_csv('NY_crime_4.csv',sep=';')
df.head()
```

```python
from folium.plugins import MarkerCluster

map = folium.Map([40.712776, -74.005974], zoom_start=13, tiles='stamen toner')

marker_cluster = MarkerCluster().add_to(map)

for i, row in df.iterrows():
    folium.Marker(location=[row['Latitude'], row['Longitude']], popup=(row['CMPLNT_FR_DT'], row['DESC']), 
                   tooltip='click me!', icon=folium.Icon(icon='warning-sign', color='red')).add_to(marker_cluster)
    
map
```
