```python
import folium
import pandas as pd
from folium import plugins

WJ_boundary = 'JawaBarat_3.geojson'
WJ_IPM = pd.read_csv('IPM2017.csv', sep=';')

WJ_IPM.head()
```

```python
m = folium.Map(location=[-6.867350, 107.604332], zoom_start=10, tiles='cartodb positron')

minimap = plugins.MiniMap()
m.add_child(minimap)


a = folium.Choropleth(geo_data=WJ_boundary, data=WJ_IPM, columns=['Kota_kab','IPM'],
                              key_on='feature.properties.Kabupaten_', fill_color='YlOrRd', fill_opacity=0.5,line_opacity=0.1,line_color='blue',
                              legend_name='Human Development Index (%)', bins=[63, 68, 73, 78, 81], highlight=True, reset=True).add_to(m)

a.geojson.add_child(folium.features.GeoJsonTooltip(fields=['Kabupaten_'], aliases=['City/Regency:'], labels=True, sticky=True))



m
```
